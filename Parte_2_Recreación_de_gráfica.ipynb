```python
#1. Creen un filtro para que el nombre de la hipoteca sea de “30 años”.  Usen len() para encontrar cuántas filas corresponden a 30 años.
filtro_mortgage = df[df['Mortgage Name'] == '30 Year']
filtro_mortgage
```

```python
# Muestro los datos de las columnas
df["Interest Rate"].value_counts()
```

```python
# Creen un filtro para una tasa de interés del 3 % (observen cómo está escrito en los datos).

tasa_interes_3porc = df[df['Interest Rate'] == 0.03]
tasa_interes_3porc
```

```python
#Definan el nuevo DataFrame basándose en los filtros de año e interés que han creado.

df1 = df[(df["Interest Rate"] == 0.03) & (df["Mortgage Name"] == "30 Year")]
df1
```

```python
filtro_tasa5 = df[df['Interest Rate'] == 0.05]
filtro_tasa5
```

```python
# Podemos usar el símbolo de negación (~) con nuestro filtro actual para darnos
#todo EXCEPTO lo que hemos definido en ese filtro. Esto funciona porque solo tenemos dos tasas de interés (cualquier valor que NO es 3 % es 5 %).

filtro_igual = (df['Interest Rate'] == 0.03) & (df['Interest Rate'] == 0.05)
filtro_negacion=df[~filtro_igual]

filtro_negacion
```

```python
# 5. Puesto que queremos graficar la suma acumulativa del interés pagado, podemos usar la función .cumsum().  ¿Cuáles son los tres primeros valores en la salida para la suma acumulativa de la tasa del 3 %?
suma_interes = df['Interest Paid'].cumsum().head(3)
suma_interes

```

```python
#Definir nuevo dataframe aplicando los filtros usados anteriormente (5%)
df2 = df[(df["Interest Rate"] == 0.05) & (df["Mortgage Name"] == "30 Year")]
df2
```

```python
#6. Usen plt.plot(x, y, c=color) para trazar la línea negra (5 %).  Queremos los valores x para que sean ['Month'] y los valores y para que sean la suma acumulativa del Interest Paid (interés pagado).
# Podemos usar el código de la pregunta anterior para definir los valores y dentro de la función plt.plot().  Pista: para el negro, c = 'k'
#Visualizar el gráfico del dataframe df1 y df2 - Mes e interés pagado
#Visualizar el gráfico del dataframe df2 - Mes e interés pagado
plt.figure(dpi=100)
plt.plot(df2["Month"],df2["Interest Paid"].cumsum(),"ko-")
plt.axis([0,300,0,400000])
plt.xlabel("Mes")
plt.ylabel("Suma acumulativa del interés pagado")
plt.show()
```

```python
#7. Para obtener ambas líneas en el mismo gráfico, solo pongan dos funciones plt.plot() una tras la otra en el mismo bloque de código y ejecútenlo.Envíen el código y gráfico reproducido. Pista: para azul, c = 'b'
#Visualizar el gráfico del dataframe df1 y df2 - Mes e interés pagado


plt.plot(df1["Month"], df1["Interest Paid"].cumsum(), 'bo-', label='30 years 3% APR')
plt.plot(df2["Month"], df2["Interest Paid"].cumsum(), 'ko-', label='30 years 5% APR')

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.axis([0, 400, 0, 500000])
plt.legend(bbox_to_anchor=(0.8, 0.2), fancybox=True, shadow=True, ncol=1)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.title('Cumulative Interest Paid')
plt.xlabel('Month')
plt.ylabel('Cumulative Interest Paid')

plt.show()

```
